```python
# Unconnected pages
import requests

url = "https://gan.wikipedia.org/w/api.php"
params = {
    "action": "query",
    "format": "json",
    "errorformat": "bc",
    "list": "querypage",
    "formatversion": 2,
    "qppage": "UnconnectedPages",
    "qplimit": "max",
}

response = requests.get(url, params=params)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    data = response.json()
    # Process the JSON response as needed
    print(data)
else:
    print(f"Error: {response.status_code}")
    print(response.text)
# Extract titles
titles = [(result['title'], result['ns']) for result in data['query']['querypage']['results']]

pages_ns0 = [title[0] for title in titles if title[1] in [10,828,12,14 ]]
```

```python
import requests
from bs4 import BeautifulSoup


def construct_sparql_query(titles):
    values_clause = "\n    ".join([f'"{title}"@zh' for title in titles])
    query = f"""
    SELECT ?lemma ?item WHERE {{
      VALUES ?lemma {{
        {values_clause}
      }}
      ?sitelink schema:about ?item;
        schema:isPartOf <https://zh.wikipedia.org/>;
        schema:name ?lemma.
    }}
    """
    return query

# # Example list of titles
file_path = 'titles.txt'  # Update with your file path
with open(file_path, 'r', encoding='utf-8') as file:
    titles = [line.strip() for line in file]
# titles = pages_ns0

# Construct SPARQL query
sparql_query = construct_sparql_query(titles)
print(sparql_query)
```

```python
import csv

# Function to convert CSV data to the desired format
def convert_csv_to_format(csv_file):
    result_list = []

    with open(csv_file, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            qid = row['item'].split('/')[-1]  # Extract QID from the item URL
            lemma = row['lemma']
            result_list.append(f'{qid}\tSganwiki\t"{lemma}"')

    return result_list

# Input CSV file
csv_file_path = 'query.csv'

# Get the result list
result = convert_csv_to_format(csv_file_path)

# Print the result
for line in result:
    print(line)
```
